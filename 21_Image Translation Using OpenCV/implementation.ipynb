```python
import cv2
import numpy as np
```

```python
img=cv2.imread('disha patnani.jpg')
img=cv2.resize(img, (500,500))

m=np.float32([[1,0,100],[0,1,50]])
new=cv2.warpAffine(img, m, (500,500))

cv2.imshow("wdf", img)
cv2.imshow("df", m)
cv2.imshow("ddf", new)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
