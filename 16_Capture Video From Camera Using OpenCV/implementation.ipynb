```python
import cv2
```

```python
cap=cv2.VideoCapture(0)
while True:
    r,frame=cap.read()
    if r==True:
        frame=cv2.resize(frame, (500,500))
        cv2.imshow('df', frame)
        if cv2.waitKey(25) & 0xff == ord('p'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
